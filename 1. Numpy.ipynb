```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image1=cv2.imread('ImageCollage dataset/top_right.jpg')
image2=cv2.imread('ImageCollage dataset/bottom_right.jpg')
image3=cv2.imread('ImageCollage dataset/top_left.jpg')
image4=cv2.imread('ImageCollage dataset/bottom_left.jpg')
image5=cv2.imread('ImageCollage dataset/center.jpeg')
```
